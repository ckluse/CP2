```python
import pandas as pd
import os
import json
import gzip
import pandas_profiling
from pathlib import Path
import re
import sys
```

```python
def load_data(file_name, head = 500):
    count = 0
    data = []
    with gzip.open(file_name) as fin:
        for l in fin:
            d = json.loads(l)
            count += 1
            data.append(d)
            
            # break if reaches the 100th line
            if (head is not None) and (count > head):
                break
    return data
```

```python
path = Path(os.getcwd()).parents[0]
```

```python
reviews = load_data(os.path.join(path,'data/source','goodreads_reviews_fantasy_paranormal.json.gz'))
inter = load_data(os.path.join(path,'data/source','goodreads_interactions_fantasy_paranormal.json.gz'))
books = load_data(os.path.join(path,'data/source','goodreads_books_fantasy_paranormal.json.gz'))
```
